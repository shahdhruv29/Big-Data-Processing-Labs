##  Process “employee.csv”

```python
%%file q1.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):                  #EmpSum that inherits from MRJob
    def mapper(self, key, line):
      record = re.split(',', line)    #splits the input line into a list called record by using a comma (,) as the delimiter.
      dno = record[2]
      salary = int(record[3])
      yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, sum(salaries)

if __name__ == '__main__':
    EmpSum.run()
```

```python
%%file q2.py
from mrjob.job import MRJob
import re

class EmpMax(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      if(record[4]=='MA'):
        dno = record[2]
        salary = int(record[3])
        yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, max(salaries)

if __name__ == '__main__':
     EmpMax.run()
```

```python
%%file q3.py
from mrjob.job import MRJob
import re
import statistics

class DepAvg(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, statistics.mean(salaries)

if __name__ == '__main__':
     DepAvg.run()
```

```python
%%file q4.py
from mrjob.job import MRJob
import re

class Emp(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      if(dno=='5' and salary>100000):
        yield "Such an employee is:- ", line


if __name__ == '__main__':
     Emp.run()
```

```python
%%file q5.py
from mrjob.job import MRJob
import re

class Gender(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      gender=record[5]
      group=(dno,gender)
      yield group,1

    def reducer(self, group,count):
        yield group, sum(count)

if __name__ == '__main__':
     Gender.run()
```

```python
# We implement Q6 a) and b)
%%file 6a.py
from mrjob.job import MRJob

class TotalReq(MRJob):

	def mapper(self, key, line):
		request = line.split(' ')
		ts = request[3]
		yield (ts[4:7],ts[8:12]), 1

	def reducer(self, ts, count):
		yield ts , sum(count)

if __name__ == '__main__':
     TotalReq.run()
```
